```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y,'x')         ## it will plot the graph points
plt.show()                ## it will show the graph representation
```

```python


x = np.array([1, 1, 1, 1, 2, 3, 4, 4, 4, 4, 3, 2,1])
y = np.array([3, 4, 5, 6, 6, 6, 6, 5, 4, 3, 3, 3,3])

plt.plot(x, y, 'x')
plt.show()
```

```python


x = np.array([1, 1, 1, 1, 2, 3, 4, 4, 4, 4, 3, 2,1])
y = np.array([3, 4, 5, 6, 6, 6, 6, 5, 4, 3, 3, 3,3])

plt.plot(x, y, marker='o')      ## marker will join the points 
plt.show()
```

```python


x = np.array([1, 1, 1, 1, 2, 3, 4, 4, 4, 4, 3, 2,1])
y = np.array([3, 4, 5, 6, 6, 6, 6, 5, 4, 3, 3, 3,3])

plt.plot(x, y, 'o:r')

plt.show()
```
