```python
##########################
# Task-3 ## give color to each fruit in pie chart
##########################

fruit = np.array([35, 25, 25, 15])
fruitlbl = ["Cherry", "Banana", "Watermellon", "Orange"]
colors = ["red", "yellow", "green", "orange"]
plt.pie(fruit, labels=fruitlbl, colors=colors)
plt.show()
```
