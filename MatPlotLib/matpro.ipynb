```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fruitList = []

while True:
    fruit = input("Enter Fruit Name")
    if fruit == "exit":
        break
    else:
        fruitList.append(fruit)

print("Fruits =",fruitList)
```

```python
count=0
for fruit in fruitList:
    count +=1
print("length of the list",count)
```

```python
fruitQtyList = []
print("Enter Quantity of fruit")
for fruit in fruitList:
    qty = int(input(f"Enter Quantity for : {fruit} "))    # f for format
    fruitQtyList.append(qty)
print("Fruits =",fruitList)
print("Quantity List: ", fruitQtyList)
```

```python
fruitClrList = []
print("Enter Color of fruit")
for fruit in fruitList:
    Clr = (input(f"Enter Color for : {fruit} "))    
    fruitClrList.append(Clr)
print("Fruits =",fruitList)
print("Color  List: ", fruitClrList)
print("Quantity List: ", fruitQtyList)
```

```python
plt.pie(fruitQtyList,labels= fruitList,colors=fruitClrList)
plt.legend(title="Fruits")
plt.show()
```

```python
maxQty = fruitQtyList[0]
maxIndex = 0

for i in range(1, len(fruitQtyList)):
    if fruitQtyList[i] > maxQty:
        maxQty = fruitQtyList[i]
        maxIndex = i
print("Max Quantity =", maxQty, "with index", maxIndex)
print("Most consumed fruit is :", fruitList[maxIndex])
```

```python
explode_value = float(input(f"Enter Explode value for {fruitList[maxIndex]} "))
explodeList = []
for i in range(len(fruitList)):
    if i == maxIndex:
        explodeList.append(explode_value)
    else:
        explodeList.append(0)
print("Explode List: ", explodeList)
```

```python
plt.pie(fruitQtyList,labels= fruitList,colors=fruitClrList,explode= explodeList)
plt.legend(title = "Fruits Consumed",loc='lower left')
plt.show()
```
