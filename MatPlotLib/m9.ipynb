```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
car= np.array(['VENUE','BENZ','AUDI','BMW'])
distance= np.array([3,8,1,10])
```

```python
plt.bar(car,distance)
plt.show()
```

```python
plt.barh(car,distance,color="orange")          # barh -> for bar in horizontal
plt.show()
```
