```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

company = (["Meta","Apple","Amazon","Netflix","Google"])
wealth = ([200000,250000,150000,300000,270000])
```

```python
plt.title("MAANG COMPANY STOCK")
plt.bar(company,wealth,color="red")
plt.show()
```

```python
## x with company name and label as MAANG , y with wealth and label as wealth(Bln) , title as MAANG COMPANY Stock , give legend as weath details
plt.title("MAANG COMPANY STOCK")
plt.xlabel("MAANG")
plt.ylabel("Wealth (Bln)")
plt.bar(company,wealth,color="red",label="Wealth Details")
plt.legend()
plt.show()
```
