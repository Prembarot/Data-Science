```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
photoupload = np.array([10,2,15,20,21,17,7,16,25,10])
likes = np.array([50,5,10,40,50,4,110,201,500,45])
plt.plot(photoupload,likes)
plt.show()
```

```python
photoupload = np.array([10,2,15,20,21,17,7,16,25,10])
likes = np.array([50,5,10,40,50,4,110,201,500,45])
plt.scatter(photoupload,likes)                  ## Scatter
plt.show()
```

```python
photoupload = np.array([10,2,15,20,21,17,7,16,25,10])
likes = np.array([50,5,10,40,50,4,110,201,500,45])
plt.plot(photoupload,likes,color='grey')
plt.scatter(photoupload,likes,color='red')                 
plt.show()
```
