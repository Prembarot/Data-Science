```python
folder = "dataset//mp"
csv_files = [f for f in os.listdir(folder) if f.endswith(".csv")]
dataframes = []
for file in csv_files:
    path = os.path.join(folder, file)
    df = pd.read_csv(path)
    df["source_file"] = file  
    dataframes.append(df)
all_sales = pd.concat(dataframes, ignore_index=True)
print("all_sales :")
print(all_sales)
```

```python
all_sales.to_csv("dataset/mp/all_sales.csv", index=False)
print("Saved as: all_sales.csv")
```
