```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "city" : ["new york","chicago","orland"],
    "temperature" : [32,14,20]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ["chicago","new york","orland"],
    "humidity" : [65,68,75]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")           ## merge --> used to merge dataframes
df3
```

```python
df1 = pd.DataFrame({
    "city" : ["chicago","new york","orland","baltimore"],
    "temperature" : [32,14,20,38]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ["chicago","new york","sand diego"],
    "humidity" : [65,68,75]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer")           ## outer join
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="inner")            ## inner join
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="left")             ## left outer join
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="right")            ## right outer join
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer",indicator=True)            ## indicator 
df3
```
