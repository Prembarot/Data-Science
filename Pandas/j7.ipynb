```python
import pandas as pd
import numpy as np
```

```python
gujarat_weather = pd.DataFrame({
    "city" : ["baroda","godhra","bharuch"],
    "temperature" : [32,40,30],
    "humidity" : [80,60,70]
})
gujarat_weather
```

```python
maharastra_weather = pd.DataFrame({
    "city" : ["mumbai","pune","nagpur"],
    "temperature" : [30,35,40],
    "humidity" : [80,60,50]
})
maharastra_weather 
```

```python
df = pd.concat([gujarat_weather,maharastra_weather ],ignore_index= True)       ## concat --> for merge data  
df                                                                             ##ignore_index= True --> it will ignore the index value and give the original numbers
```

```python
df = pd.concat([gujarat_weather,maharastra_weather ],keys=["Guj","Maha"])        ## Key will give the main column name 
df 
```
